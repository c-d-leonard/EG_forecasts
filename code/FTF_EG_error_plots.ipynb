```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ELGs_baseline_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI_4MOST_ELGs.txt')
ELGs_baseline_area = 4000

ELGs_150pc_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI150pc_4MOST_ELGs.txt')
ELGs_150pc_area = 5500

ELGs_200pc_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI200pc_4MOST_ELGs.txt')
ELGs_200pc_area = 7000

ELGs_LSSTarea_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI_4MOST_18000deg2_ELGs.txt')
ELGs_LSSTarea_area = 18000
```

```python
ELG_errors = [ELGs_baseline_err, ELGs_150pc_err, ELGs_200pc_err, ELGs_LSSTarea_err]
ELG_areas = [ELGs_baseline_area, ELGs_150pc_area, ELGs_200pc_area, ELGs_LSSTarea_area]
```

```python
plt.figure()
plt.rcParams["font.family"] = "serif"
plt.plot(ELG_areas, ELG_errors, 'ko')
plt.xlabel('Areas, deg$^2$', fontsize=16)
plt.ylabel('$\sigma(E_G)$, z=1.0', fontsize=16)
plt.tick_params(axis='both', labelsize='14')
plt.tight_layout()
plt.savefig('./EG_err_ELGs.png')
```

```python
LRGs_baseline_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI_4MOST_LRGs.txt')
LRGs_baseline_area = 10500
LRGs_150pc_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI150pc_4MOST_LRGs.txt')
LRGs_150pc_area = 12000
LRGs_200pc_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI200pc_4MOST_LRGs.txt')
LRGs_200pc_area = 13500
LRGs_LSSTarea_err = np.loadtxt('/home/danielle/Documents/CMU/Research/EG_comparison/txtfiles/Eg_err_DESI_4MOST_18000deg2_LRGs.txt')
LRGs_LSSTarea_area = 18000
```

```python
LRG_errors = [LRGs_baseline_err, LRGs_150pc_err, LRGs_200pc_err, LRGs_LSSTarea_err]
LRG_areas = [LRGs_baseline_area, LRGs_150pc_area, LRGs_200pc_area, LRGs_LSSTarea_area]
```
