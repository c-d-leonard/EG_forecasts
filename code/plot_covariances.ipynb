```python
import numpy as np
import matplotlib.pyplot as plt
```

This is to compare the covariance as calculated assuming covariance values at rp0 can be approximated as the bin averaged over rp0 with that which tries to get the more exact rp0 value.

```python
cov_rp0exact = np.loadtxt('../txtfiles/joint_covariance_rp0exactinterp_Aug23_2024.dat')
cov_rp0approx = np.loadtxt('../txtfiles/joint_covariance_rp0=rp_edges[0]_Aug21_2024.dat')
```

```python
plt.figure(figsize=[8,8])
plt.imshow(np.log10(cov_rp0exact))
cbar1=plt.colorbar()
plt.imshow(np.log10(-cov_rp0exact), cmap=plt.cm.magma)
cbar2=plt.colorbar()
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
cbar1.ax.tick_params(labelsize=15)
cbar2.ax.tick_params(labelsize=15)
cbar1.set_label('$\log_{10}(Cov)$', fontsize=15)
cbar2.set_label('$\log_{10}(-Cov)$', fontsize=15)
#plt.savefig('../plots/hybrid_cov_log10.pdf')
plt.show()
```

```python
plt.figure(figsize=[8,8])
plt.imshow(np.log10(cov_rp0approx))
cbar1=plt.colorbar()
plt.imshow(np.log10(-cov_rp0approx), cmap=plt.cm.magma)
cbar2=plt.colorbar()
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
cbar1.ax.tick_params(labelsize=15)
cbar2.ax.tick_params(labelsize=15)
cbar1.set_label('$\log_{10}(Cov)$', fontsize=15)
cbar2.set_label('$\log_{10}(-Cov)$', fontsize=15)
#plt.savefig('../plots/hybrid_cov_log10.pdf')
plt.show()
```

```python
plt.figure(figsize=[8,8])
plt.imshow(np.log10(np.abs(cov_rp0approx - cov_rp0exact)/np.abs(cov_rp0exact)))
cbar1=plt.colorbar()
#plt.imshow(np.log10(-cov_rp0approx), cmap=plt.cm.magma)
#cbar2=plt.colorbar()
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
cbar1.ax.tick_params(labelsize=15)
#cbar2.ax.tick_params(labelsize=15)
cbar1.set_label('$\log_{10}(Cov)$', fontsize=15)
#cbar2.set_label('$\log_{10}(-Cov)$', fontsize=15)
#plt.savefig('../plots/hybrid_cov_log10.pdf')
plt.show()
```
