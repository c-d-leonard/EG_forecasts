This notebook is going to deal with getting the appropriate posterior for the scale-dependent bias paramters to sample from and feed into our correction to E_G.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from getdist.gaussian_mixtures import GaussianND
from getdist import plots
```

```python
# We have mean values of b1_L and b2_L from Kitandis & White 2022 for a DESI-LRG-like sample. These are labeled as such because they are Lagrangian bias pars.
# We will switch to Eulerian before doing our calculation.

b1_L_mean = 1.333
b2_L_mean = 0.514
means = [b1_L_mean, b2_L_mean]

# We have estimated the covariance visually from their plot (with a little help from chatgpt).

cov = np.zeros((2,2))
cov[0,0] = 0.0016
cov[1,1] = 0.0650
cov[0,1] = -0.005
cov[1,0] = -0.005

# Let's set this up in a PDF and make sure it looks right when plotted:

labels = ['b1L', 'b2L']
names =['b1L', 'b2L']

dist_bias = GaussianND(means, cov, labels = labels, names = names)
                       
settings = plots.GetDistPlotSettings()
settings.fontsize=6
settings.axes_fontsize = 6
    
# Make plot
dist_plot = plots.get_subplot_plotter(width_inch=3)
dist_plot.triangle_plot([dist_bias], filled = True)
```
